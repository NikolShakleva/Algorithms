```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt


random = pd.read_excel("python_random.xlsx", sheet_name= "BS_random")
```

```python
random_k10 = random.loc[random['K']== 10]
```

```python
# random
binary_avg = pd.read_excel("python_random.xlsx", sheet_name= "BS_random")
binary_avg = binary_avg[["AVG", "N_Seed1"]]
binary_avg = binary_avg.dropna()

tabulation_avg = pd.read_excel("python_random.xlsx", sheet_name= "T_random")
tabulation_avg = tabulation_avg[["AVG", "N_Seed1"]]
tabulation_avg = tabulation_avg.dropna()

tabulation_5 = pd.read_excel("python_random.xlsx", sheet_name= "T_random")
tabulation_5 = tabulation_5[["AVG", "N_Seed1", "K"]]
tabulation_5 = tabulation_5.loc[tabulation_5['K']== 15]
tabulation_5 = tabulation_5.dropna()

```

```python
#RANDOM
b_x = binary_avg["N_Seed1"].values
b_y = binary_avg["AVG"].values

t_x = tabulation_avg["N_Seed1"].values
t_y = tabulation_avg["AVG"].values


# t5_x = tabulation_5["N_Seed1"].values
# t5_y = tabulation_5["AVG"].values


plt.plot(b_x, b_y, 'r', label = "BinarySearch")
plt.plot(t_x, t_y, 'b', label = "Tabulation")
# plt.plot(t5_x, t5_y, 'g', label = "k5")

plt.xlabel('N')
plt.ylabel('running time')
plt.title('Mode: Random with k = 10')
plt.legend()
plt.show()
```

```python
# non-existent
b_nonExistent = pd.read_excel("python_nonExistent.xlsx", sheet_name= "BS_nonExistent")
b_nonExistent = b_nonExistent[["AVG", "N_Seed1"]]
b_nonExistent = b_nonExistent.dropna()

t_nonExistent = pd.read_excel("python_nonExistent.xlsx", sheet_name= "T_nonExistent")
t_nonExistent = t_nonExistent[["AVG", "N_Seed1"]]
t_nonExistent = t_nonExistent.dropna()
```
